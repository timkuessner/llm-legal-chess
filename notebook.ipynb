```python
import chess
import random
```

```python
board = chess.Board()

while not board.is_game_over():
    legal_moves = [m.uci() for m in board.legal_moves]
    print("FEN:\n" + str(board))
    print("Legal:" + str(legal_moves))

    move = random.choice(legal_moves)
    board.push_uci(move)
```
